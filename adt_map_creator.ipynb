```python
import json

import pandas as pd
```

```python
with open(r'C:\Users\cmullins\Desktop\ksi\site_adts.txt') as f:
    site_adts_df = json.loads(f.read())

with open(r'C:\Users\cmullins\Desktop\ksi\site_locs.txt') as f:
    site_locs_df = json.loads(f.read())
```

```python
site_adts = pd.DataFrame(site_adts_df['data'])
site_locs = pd.DataFrame(site_locs_df['data']).T.reset_index(drop=True)
```

```python
df = pd.merge(site_adts, site_locs, left_on='siteId', right_on='id')
df.head()
```

```python
locs = df['location'].apply(lambda x: pd.Series(x))
```

```python
df = pd.concat([df, locs], axis=1)
df = df.filter([
    'siteId',
    'node',
    'cosit',
    'description',
    'year',
    'aadt',
    'lat',
    'lng'
])
```

```python
df['hyperlink'] = 'https://drakewell02.drakewell.com/calendar_alt.asp?node=' + df['node'] + '&cosit=' + df['cosit']
```

```python
df.dropna(subset='aadt') \
    .to_csv('site_adts.csv', index=False)
```
